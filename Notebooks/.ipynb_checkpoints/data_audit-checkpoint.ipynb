## Volume and scale (from overlay plot)
FOODS completely dominates — averaging 20,000–30,000 daily units across all stores, roughly 6–7× higher than HOUSEHOLD (~5,000–10,000) and 8–10× higher than HOBBIES (~3,000–4,000). This volume gap means a single FOODS model error has outsized impact on the overall WRMSSE score. FOODS needs the most careful feature engineering.
Upward trend across all three categories (from overlay plot)
All three categories grow consistently from 2011 to 2016. FOODS grows from ~18,000 to ~30,000 daily units — roughly 65% over 5 years. HOUSEHOLD nearly doubles from ~5,000 to ~10,000. HOBBIES grows modestly from ~3,000 to ~4,000. This trend must be captured — a model trained only on early years will systematically under-forecast later periods. A time index or year feature is non-negotiable.
## Weekly seasonality is strong and consistent (from 3-panel plot)
The raw daily line shows a tight sawtooth pattern throughout all 5 years in all three categories. The spikes repeat every 7 days — weekend sales are visibly higher. This confirms that wday is one of the most important features to include, and lag features at 7-day intervals will be highly predictive.
The shaded holiday windows expose a surprising pattern (from 3-panel plot)
The Dec–Jan shading shows FOODS sales actually drop sharply at Christmas — visible as the near-zero dips in 2012, 2013, 2014, 2015, 2016. This matches the outlier detection result — Christmas Day is when stores close, not a high-sales day. HOBBIES and HOUSEHOLD show the same Christmas drop. Counterintuitively, the holiday season peak for all categories happens before Christmas, not on it.
## 2013 HOBBIES anomaly (from 3-panel plot)
There is a clear structural break in HOBBIES in early 2013 — sales jump from ~2,500 to ~3,500 daily units and hold at the new level. This is not a seasonal spike, it's a step change. Likely explanation: new stores opened or a new product range was listed. Worth flagging for the model — a store-count or assortment-change feature could capture this if external data is available.
## 2016 seasonality heatmap is incomplete (from heatmap)
The 2016 row cuts off at May — June through December are black/empty. This is because the validation file only goes to mid-2016. This is exactly why Bug 5 matters — loading sales_train_evaluation.csv instead gives you data through June 2016, which is still incomplete for the full year but correct for what the competition provides. Do not read the black cells as zero sales — they are missing data.
Seasonality heatmap shows mid-year peaks, not year-end (from heatmap)
Contrary to what you might expect from a retail dataset, FOODS peaks appear in Jun–Sep across 2012–2015, not in November–December. The 2013 Aug–Sep block is the darkest cell on the entire heatmap. This could reflect summer SNAP distribution patterns or warmer-weather food purchasing behaviour. This is a genuine insight that tells you a simple "Q4 is peak" assumption would be wrong for this dataset — month and season features need to be learned from data, not hardcoded.

```python

# ── Cap outliers using IQR per category on daily aggregated sales ──
# Using daily_cat (aggregated) not sales_long (raw item-day)
# because raw item-day is mostly zeros which distorts percentiles

daily_cat["sales_capped"] = daily_cat["sales"].copy()

for cat in daily_cat["cat_id"].unique():
    mask = daily_cat["cat_id"] == cat
    data = daily_cat.loc[mask, "sales"]

    Q1  = data.quantile(0.25)
    Q3  = data.quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    daily_cat.loc[mask, "sales_capped"] = data.clip(lower=lower, upper=upper)

```

```python
print("── Capping Summary ──\n")
for cat in daily_cat["cat_id"].unique():
    subset = daily_cat[daily_cat["cat_id"] == cat]
    print(f"{cat}")
    print(f"  Raw    — max: {subset['sales'].max():.0f}  "
          f"mean: {subset['sales'].mean():.1f}")
    print(f"  Capped — max: {subset['sales_capped'].max():.0f}  "
          f"mean: {subset['sales_capped'].mean():.1f}\n")
```

```python
# Create a single SNAP flag per row ─────────────────
# Each store belongs to one state — snap flag depends on that state
# Vectorised: no apply(), no loop

sales_long["is_snap"] = (
    ((sales_long["state_id"] == "CA") & (sales_long["snap_CA"] == 1)) |
    ((sales_long["state_id"] == "TX") & (sales_long["snap_TX"] == 1)) |
    ((sales_long["state_id"] == "WI") & (sales_long["snap_WI"] == 1))
).astype(int)
```

```python
# Drop individual state snap cols — replaced by is_snap
sales_long.drop(columns=["snap_CA", "snap_TX", "snap_WI"], inplace=True)

print("SNAP flag value counts:")
print(sales_long["is_snap"].value_counts(), "\n")
```

```python
# 1. reduce memory first
for col in ['store_id', 'item_id']:
    sales_long[col] = sales_long[col].astype('category')
    price_df[col] = price_df[col].astype('category')

sales_long['wm_yr_wk'] = sales_long['wm_yr_wk'].astype('int16')
price_df['wm_yr_wk'] = price_df['wm_yr_wk'].astype('int16')

price_df['sell_price'] = price_df['sell_price'].astype('float32')

# 2. drop duplicates (VERY important)
price_df = price_df.drop_duplicates(['store_id', 'item_id', 'wm_yr_wk'])

# 3. merge
sales_long = sales_long.merge(
    price_df,
    on=["store_id", "item_id", "wm_yr_wk"],
    how="left"
)
```

```python
#  Check nulls introduced by merges ──────────────────
print("\n── Post-merge null check ──")
null_check = sales_long[["date", "sell_price", "is_snap"]].isnull().sum()
print(null_check)

price_null_pct = sales_long["sell_price"].isnull().mean() * 100
print(f"\nsell_price null: {price_null_pct:.2f}% of rows")
print("→ Expected — items have no price before they were listed in stores")
```

```python
# Forward-fill missing prices within each item-store──
# Price doesn't change mid-week, so last known price is a safe fill
sales_long["sell_price"] = (
    sales_long
    .sort_values(["item_id", "store_id", "date"])
    .groupby(["item_id", "store_id"], observed=True)["sell_price"]
    .ffill()
)
```

```python
# Check nulls remaining after ffill
remaining_nulls = sales_long["sell_price"].isnull().sum()
print(f"\nsell_price nulls after ffill: {remaining_nulls}")
print("→ Remaining nulls are items with NO price record at all (pre-launch rows)")
```

```python
#  Add helper flags ──────────────────────────────────
sales_long["is_event"]   = sales_long["event_name_1"].notna().astype(int)
sales_long["is_weekend"] = sales_long["wday"].isin([1, 2]).astype(int)
# wday: 1=Saturday, 2=Sunday in M5 calendar
```

```python
print("\n── Final merged dataframe ──")
print(f"  Shape   : {sales_long.shape}")
print(f"  Columns : {list(sales_long.columns)}")
print(f"\n  Sample rows:")
print(sales_long[["item_id", "store_id", "date", "sales",
                   "sell_price", "is_snap", "is_event",
                   "is_weekend", "month", "wday"]].head(5).to_string(index=False))
```
